```python
import pandas as pd
import numpy as np
from numpy.linalg import inv
```

```python
data = pd.read_excel("SETHD.xlsx")
nameStock = list(data.iloc[:,0])
dataDimension = data.shape
```

```python
data=data.iloc[0:dataDimension[0],1:dataDimension[1]]
```

```python
data_array=data.as_matrix()
```

```python
mean=[]
for i in range(len(data_array)):
    mean.append((data_array[i].mean())*100)
```

```python
zig_array = np.full((dataDimension[0], dataDimension[0]), np.NaN)
for i in range(len(data_array)):
    for j in range(len(data_array)):
        zig_array[i][j]=(np.cov(data_array[i].astype(float),data_array[j].astype(float),rowvar=False)[0][1])*10000
```

```python
result = np.full((2+dataDimension[0], 2+dataDimension[0]), np.NaN)
```

```python
result[0][0]=0
result[0][1]=0
result[1][0]=0
result[1][1]=0

for i in range(2,len(result)):
    result[1][i]=1
    result[i][1]=1
    
    result[0][i]=mean[i-2]
    result[i][0]=mean[i-2]
for i in range(2,len(result)):
    for j in range(2,len(result)):
        result[i][j]=zig_array[i-2][j-2]
```

```python
pd.DataFrame(result)
```

```python
u = np.array([1]*dataDimension[0])
```

```python
# หาอินเวิร์ส
zig_inv=inv(zig_array)
```

```python
top = np.matmul(u,zig_inv)
```

```python
bottom = np.matmul(u,zig_inv)
bottom = np.matmul(bottom,np.transpose(u))
```

```python
bottom
```

```python
result = top/bottom
```
