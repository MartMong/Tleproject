```python
import pandas as pd
import numpy as np
from numpy.linalg import inv
import itertools
```

# Prepare data

```python
data = pd.read_excel("16-ตัวให้มง.xlsx")
nameStock = list(data.iloc[:,0])
dataDimension = data.shape
```

```python
data=data.iloc[0:dataDimension[0],1:dataDimension[1]]
```

```python
data_array_input=data.as_matrix()
```

```python
rValue =10

combi = list(itertools.combinations(list(range(dataDimension[0])), rValue))
```

```python
def findResult(data,combi):
    result = []
    c=0
#  retrive data
    for case in combi:
        data_array=data_array_input[list(case)]
        
        mean=[]
        for i in range(len(data_array)):
            mean.append((data_array[i].mean()))

        zig_array = np.full((dataDimension[0], dataDimension[0]), np.NaN)
        for i in range(len(data_array)):
            for j in range(len(data_array)):
                zig_array[i][j]=(np.cov(data_array[i].astype(float),data_array[j].astype(float),rowvar=False)[0][1])

        # Find H Array
        H = np.full((2+rValue, 2+rValue), np.NaN)
        H[0][0]=0
        H[0][1]=0
        H[1][0]=0
        H[1][1]=0

        for i in range(2,len(H)):
            H[1][i]=1
            H[i][1]=1

            H[0][i]=mean[i-2]
            H[i][0]=mean[i-2]
        for i in range(2,len(H)):
            for j in range(2,len(H)):
                H[i][j]=zig_array[i-2][j-2]

        # H inverse operation

        
        if np.linalg.det(H)!=0:
            H_inv=inv(H)
            
            ERp=6
            erpArray = [ERp,1] +[0]*(rValue)

            H_operation_result=np.matmul(H_inv,erpArray)
            ten_result = H_operation_result[2:]
            if all(x > 0 for x in ten_result):
                result.append(case)
            print(ten_result)
    return result
```

```python
result = findResult(data_array_input,combi)
```
